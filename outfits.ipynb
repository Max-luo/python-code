```python
from astropy.io import fits
import numpy as np
DR14Q_v4_4 = fits.open('./DR14Q_v4_4.fits')
DR14Q_v4_4.info()
data = DR14Q_v4_4[1].data
```

```python
mask = ((data.field(8)>1.2) & (data.field(8)<1.5) & (data.field(47)>0))
data = data[mask]
```

```python
c0 = fits.Column(name='SDSS_NAME', array=data.field(0), format='18a')
c1 = fits.Column(name='Z', array=data.field(8), format='D')
c2 = fits.Column(name='Z_ERR', array=data.field(9), format='D')
c3 = fits.Column(name='PSFFLUX', array=data.field(38), format='5D')
c4 = fits.Column(name='PSFMAG', array=data.field(40), format='5D')
c5 = fits.Column(name='ERR_PSFMAG', array=data.field(41), format='5D')
c6 = fits.Column(name='GAL_EXT', array=data.field(43), format='5D')
c7 = fits.Column(name='FLUX_0.2_2.0keV', array=data.field(47), format='D')
c8 = fits.Column(name='FLUX_0.2_2.0keV_ERR', array=data.field(48), format='D')
t = fits.BinTableHDU.from_columns([c0, c1, c2, c3, c4, c5, c6, c7, c8])
t.writeto('brief.fits')
```

```python
def Namesindex(Name):
    for i in range(len(data.names)):
        if Name == data.names[i]:
            return i
```
